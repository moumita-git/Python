```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'your_file.xlsx' with the actual path to your Excel file
file_path ="Data Analysis - Caselet#2.xlsx"

# Replace 'different_format' with the actual sheet name in your Excel file
sheet_name = 'DifferentFormat'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Create a box plot for the 'TPT' column
df.boxplot(column='TPT',by='Team Member')
# Set labels and title
plt.xlabel('Team Member')
plt.ylabel('Day TPT')
plt.xticks(rotation=90)

plt.title('Box Plot of TPT for Team Members')

# Show the plot
plt.show()
```
